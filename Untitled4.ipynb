```python
circuit = QuantumCircuit(3, 3)
circuit.name = "Qiskit Sample - 3-qubit GHZ circuit"
circuit.h(0)
circuit.cx(0, 1)
circuit.cx(1, 2)
circuit.measure([0,1,2], [0, 1, 2])

# Print out the circuit
circuit.draw()
```

```python
provider = IBMQ.get_provider("ibm-q")

for backend in provider.backends():
  try:
    qubit_count = len(backend.properties().qubits)
  except:
    qubit_count = "simulated"

  print(f"{backend.name()} has {backend.status().pending_jobs} queued and {qubit_count} qubits")

from qiskit.tools.monitor import job_monitor
backend = provider.get_backend("ibmq_qasm_simulator")
job = q.execute(circuit, backend=backend, shots=500)
job_monitor(job)
```

```python
from qiskit.visualization import plot_histogram

result = job.result()
counts = result.get_counts(circuit)

plot_histogram([counts])
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram
```

```python
statevector_sim = q.Aer.get_backend("statevector_simulator")
qasm_sim = q.Aer.get_backend("qasm_simulator")
```

```python
def run_job(circuit):
  result = q.execute(circuit, backend=statevector_sim).result()
  statevec = result.get_statevector()

  n_qubits = circuit.n_qubits
  circuit.measure([i for i in range(n_qubits)], [i for i in range(n_qubits)])

  qasm_job = q.execute(circuit, backend=qasm_sim, shots=1024).result()
  counts = qasm_job.get_counts()
  return statevec, counts
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

state = Statevector.from_instruction(qc)
plot_bloch_multivector(state, title="New Bloch Multivector", reverse_bits=False)
```
